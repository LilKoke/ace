```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
omega_df = pd.read_csv("out/assignment2/omega.csv", sep=" ")
q_df = pd.read_csv("out/assignment2/q.csv", sep=" ")
domega_df = pd.read_csv("out/assignment2/domega.csv", sep=" ")
dq_df = pd.read_csv("out/assignment2/dq.csv", sep=" ")
omega_s_df = pd.read_csv("out/assignment2/omega_s.csv", sep=" ")
q_s_df = pd.read_csv("out/assignment2/q_s.csv", sep=" ")
P_df = pd.read_csv("out/assignment2/P.csv", sep=" ")
```

```python
t_omega = omega_df.iloc[:,0].values
omega1 = omega_df.iloc[:,1].values
omega2 = omega_df.iloc[:,2].values
omega3 = omega_df.iloc[:,3].values
```

```python
t_q = q_df.iloc[:,0].values
q0 = q_df.iloc[:,1].values
q1 = q_df.iloc[:,2].values
q2 = q_df.iloc[:,3].values
q3 = q_df.iloc[:,4].values
```

```python
t_domega = domega_df.iloc[:,0].values
domega1 = domega_df.iloc[:,1].values
domega2 = domega_df.iloc[:,2].values
domega3 = domega_df.iloc[:,3].values
```

```python
t_dq = dq_df.iloc[:,0].values
dq0 = dq_df.iloc[:,1].values
dq1 = dq_df.iloc[:,2].values
dq2 = dq_df.iloc[:,3].values
dq3 = dq_df.iloc[:,4].values
```

```python
t_omega_s = omega_s_df.iloc[:,0].values
omega_s1 = omega_s_df.iloc[:,1].values
omega_s2 = omega_s_df.iloc[:,2].values
omega_s3 = omega_s_df.iloc[:,3].values
```

```python
t_q_s = q_s_df.iloc[:,0].values
q_s0 = q_s_df.iloc[:,1].values
q_s1 = q_s_df.iloc[:,2].values
q_s2 = q_s_df.iloc[:,3].values
q_s3 = q_s_df.iloc[:,4].values
```

```python
t_q_s = q_s_df.iloc[:,0].values
Pq0 = P_df.iloc[:,1].values
Pq1 = P_df.iloc[:,2].values
Pq2 = P_df.iloc[:,3].values
Pq3 = P_df.iloc[:,4].values
Pomega1 = P_df.iloc[:,5].values
Pomega2 = P_df.iloc[:,6].values
Pomega3 = P_df.iloc[:,7].values
```

```python
fig = plt.figure(figsize=(14.4, 4.8))
ax1 = fig.add_subplot(121)
ax1.plot(t_omega, omega1, label="$\omega_1$") 
ax1.plot(t_omega, omega2, label="$\omega_2$")
ax1.plot(t_omega, omega3, label="$\omega_3$")
ax1.set_xlabel("$t$ [s]")
ax1.set_ylabel("$\omega$ [rad/s]")
ax1.grid(axis='y')
ax1.set_title("$\omega_t$")
ax1.legend()
ax2 = fig.add_subplot(122)
ax2.plot(t_q, q0, label="$q_0$")
ax2.plot(t_q, q1, label="$q_1$")
ax2.plot(t_q, q2, label="$q_2$")
ax2.plot(t_q, q3, label="$q_3$")
ax2.grid(axis='y')
ax2.set_xlabel("$t$ [s]")
ax2.set_ylabel("$q$")
ax2.set_title("$q_t$")
ax2.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(14.4, 4.8))
ax1 = fig.add_subplot(121)
ax1.set_title("$\omega_t - \omega_s$")
ax1.plot(t_domega, domega1, label="$\omega_1$") 
ax1.plot(t_domega, domega2, label="$\omega_2$")
ax1.plot(t_domega, domega3, label="$\omega_3$")
ax1.set_xlabel("$t$ [s]")
ax1.set_ylabel("$\omega$ [rad/s]")
ax1.grid(axis='y')
ax1.legend()
ax2 = fig.add_subplot(122)
ax2.set_title("$q_t - q_s$")
ax2.plot(t_dq, dq0, label="$q_0$")
ax2.plot(t_dq, dq1, label="$q_1$")
ax2.plot(t_dq, dq2, label="$q_2$")
ax2.plot(t_dq, dq3, label="$q_3$")
ax2.grid(axis='y')
ax2.set_xlabel("$t$ [s]")
ax2.set_ylabel("$q$")
ax2.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(14.4, 4.8))
ax1 = fig.add_subplot(121)
ax1.plot(t_omega_s, omega_s1, label="$\omega_1$") 
ax1.plot(t_omega_s, omega_s2, label="$\omega_2$")
ax1.plot(t_omega_s, omega_s3, label="$\omega_3$")
ax1.set_xlabel("$t$ [s]")
ax1.set_ylabel("$\omega$ [rad/s]")
ax1.grid(axis='y')
ax1.set_title("$\omega_s$")
ax1.legend()
ax2 = fig.add_subplot(122)
ax2.plot(t_q_s, q_s0, label="$q_0$")
ax2.plot(t_q_s, q_s1, label="$q_1$")
ax2.plot(t_q_s, q_s2, label="$q_2$")
ax2.plot(t_q_s, q_s3, label="$q_3$")
ax2.grid(axis='y')
ax2.set_xlabel("$t$ [s]")
ax2.set_ylabel("$q$")
ax2.set_title("$q_s$")
ax2.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(14.4, 14.4))
ax1 = fig.add_subplot(421)
ax1.plot(t_dq, dq0, label="$q_0$")
ax1.plot(t_dq, Pq0, label="$P$")
ax1.plot(t_dq, -Pq0, label="$-P$")
ax1.set_ylim(-0.1, 0.1)
ax1.legend()
ax2 = fig.add_subplot(422)
ax2.plot(t_dq, dq1, label="$q_1$")
ax2.plot(t_dq, Pq1, label="$P$")
ax2.plot(t_dq, -Pq1, label="$-P$")
ax2.set_ylim(-0.05, 0.05)
ax2.legend()
ax3 = fig.add_subplot(423)
ax3.plot(t_dq, dq2, label="$q_2$")
ax3.plot(t_dq, Pq2, label="$P$")
ax3.plot(t_dq, -Pq2, label="$-P$")
ax3.set_ylim(-0.05, 0.05)
ax3.legend()
ax4 = fig.add_subplot(424)
ax4.plot(t_dq, dq3, label="$q_3$")
ax4.plot(t_dq, Pq3, label="$P$")
ax4.plot(t_dq, -Pq3, label="$-P$")
ax4.set_ylim(-0.05, 0.05)
ax4.legend()
ax5 = fig.add_subplot(425)
ax5.plot(t_domega, domega1, label="$\omega_1$")
ax5.plot(t_dq, Pomega1, label="$P$")
ax5.plot(t_dq, -Pomega1, label="$-P$")
ax5.set_ylim(-0.1, 0.1)
ax5.legend()
ax6 = fig.add_subplot(426)
ax6.plot(t_domega, domega2, label="$\omega_2$")
ax6.plot(t_dq, Pomega2, label="$P$")
ax6.plot(t_dq, -Pomega2, label="$-P$")
ax6.set_ylim(-0.025, 0.025)
ax6.legend()
ax7 = fig.add_subplot(427)
ax7.plot(t_domega, domega3, label="$\omega_3$")
ax7.plot(t_dq, Pomega3, label="$P$")
ax7.plot(t_dq, -Pomega3, label="$-P$")
ax7.set_ylim(-0.1, 0.1)
ax7.legend()
plt.show()
```
