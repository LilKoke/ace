```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
omega_df = pd.read_csv("out/assignment2/omega.csv", sep=" ")
q_df = pd.read_csv("out/assignment2/q.csv", sep=" ")
domega_df = pd.read_csv("out/assignment2/domega.csv", sep=" ")
dq_df = pd.read_csv("out/assignment2/dq.csv", sep=" ")
```

```python
t_omega = omega_df.iloc[:,0].values
omega1 = omega_df.iloc[:,1].values
omega2 = omega_df.iloc[:,2].values
omega3 = omega_df.iloc[:,3].values
```

```python
t_q = q_df.iloc[:,0].values
q0 = q_df.iloc[:,1].values
q1 = q_df.iloc[:,2].values
q2 = q_df.iloc[:,3].values
q3 = q_df.iloc[:,4].values
```

```python
t_domega = domega_df.iloc[:,0].values
domega1 = domega_df.iloc[:,1].values
domega2 = domega_df.iloc[:,2].values
domega3 = domega_df.iloc[:,3].values
```

```python
t_dq = dq_df.iloc[:,0].values
dq0 = dq_df.iloc[:,1].values
dq1 = dq_df.iloc[:,2].values
dq2 = dq_df.iloc[:,3].values
dq3 = dq_df.iloc[:,4].values
```

```python
fig = plt.figure(figsize=(14.4, 4.8))
ax1 = fig.add_subplot(121)
ax1.plot(t_omega, omega1, label="$\omega_1$") 
ax1.plot(t_omega, omega2, label="$\omega_2$")
ax1.plot(t_omega, omega3, label="$\omega_3$")
ax1.legend()
ax2 = fig.add_subplot(122)
ax2.plot(t_q, q0, label="$q_0$")
ax2.plot(t_q, q1, label="$q_1$")
ax2.plot(t_q, q2, label="$q_2$")
ax2.plot(t_q, q3, label="$q_3$")
ax2.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(14.4, 4.8))
ax1 = fig.add_subplot(121)
ax1.plot(t_domega, domega1, label="$\omega_1$") 
ax1.plot(t_domega, domega2, label="$\omega_2$")
ax1.plot(t_domega, domega3, label="$\omega_3$")
ax1.legend()
ax2 = fig.add_subplot(122)
ax2.plot(t_dq, dq0, label="$q_0$")
ax2.plot(t_dq, dq1, label="$q_1$")
ax2.plot(t_dq, dq2, label="$q_2$")
ax2.plot(t_dq, dq3, label="$q_3$")
ax2.legend()
plt.show()
```
